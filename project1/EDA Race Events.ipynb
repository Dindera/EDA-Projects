```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
import plotly.express as px

plt.style.use('fivethirtyeight')
pd.set_option('display.max_columns', 500)
color_pal = plt.rcParams["axes.prop_cycle"].by_key()["color"]
```

```python
# Get the data
df = pd.read_csv('Data/TWO_CENTURIES_OF_UM_RACES.csv')
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.dtypes
```

```python
# Clean up Data
```

```python
# Only get UK Races, 50k, 2020

# Show 50mi or 50k data
df[df['Event distance/length'] == '50mi']
```

```python
# combine 50km/50mi with isin

df[df['Event distance/length'].isin(['50km', '50mi'])]
```

```python
df[(df['Event distance/length'].isin(['50km', '50mi'])) & df['Year of event'].isin([2020, 2021, 2022])]
```

```python
df[df['Event name'] == 'Everglades 50 Mile Ultra Run (USA)']['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
# combine filters together
df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'].isin([2020, 2021, 2022])) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2 = df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'].isin([2020, 2021, 2022])) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.shape
```

```python
# Remove (USA) from event name

df2['Event name'].str.split('(').str.get(0)
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
# Clean the athlete age
df2['Athlete age'] = df2['Year of event'] - df2['Athlete year of birth']
```

```python
# Remove h from Athlete performance

df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head(3)
```

```python
# drop columns: Athlete Club, Athlete Country, Athlete year of birth, Athlete Age Category

df2 = df2.drop(['Athlete club','Athlete country', 'Athlete year of birth', 'Athlete age category'], axis = 1)
```

```python
df2.head(3)
```

```python
# clean null values

df2.isna().sum()
```

```python
df2[df2['Athlete age'].isna() == 1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
#check for duplicates
df2[df2.duplicated() == True]
```

```python
#reset index
df2.reset_index(drop = True)
```

```python
df2.dtypes
```

```python
df2['Athlete age'] = df2['Athlete age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
df2.head()
```

```python
df2 = df2.rename(columns= {
                'Year of event': 'year',                  
                'Event dates': 'race_day',              
                'Event name': 'race_name',                    
                'Event distance/length': 'race_length',        
                'Event number of finishers': 'race_number_of_finisher',      
                'Athlete performance': 'athlete_performance',           
                'Athlete gender': 'athlete_gender',                
                'Athlete average speed': 'athlete_average_speed',         
                'Athlete ID': 'athlete_id',                   
                'Athlete age': 'athlete_age'
                })
```

```python
df2.head(5)
```

```python
#reset index
df2.reset_index(drop = True)
```

```python
df3 = df2[['race_day', 'race_name', 'race_length', 'race_number_of_finisher', 'athlete_id', 'athlete_gender', 'athlete_age', 'athlete_performance', 'athlete_average_speed', 'year' ]]
```

```python
df3.head(3)
```

```python
# Plot race length

sns.histplot(df3['race_length'])
```

```python
sns.histplot(df3, x = 'race_length', hue='athlete_gender')
```

```python
sns.displot(df3[df3['race_length'] == '50mi']['athlete_average_speed'])
```

```python
sns.violinplot(data = df3, x='race_length', y='athlete_average_speed', hue='athlete_gender', inner ='quart', linewidth=1)
```

```python
sns.lmplot(data=df3, x='athlete_age', y='athlete_average_speed', hue='athlete_gender')
```

```python
# Find the difference in speed for the 50km, 50mi male, female, X
df3.groupby(['race_length', 'athlete_gender'])['athlete_average_speed']
   .mean()
```

```python
#Age groups best in the 50m Race 
df3.query('race_length == "50mi"')\
    .groupby('athlete_age')['athlete_average_speed']\
    .agg(['mean', 'count'])\
    .sort_values('mean', ascending=False)\
    .query('count>19')
```

```python
#Age groups worst in the 50m Race 
df3.query('race_length == "50mi"')\
    .groupby('athlete_age')['athlete_average_speed']\
    .agg(['mean', 'count'])\
    .sort_values('mean', ascending=True)\
    .query('count>9')
```

```python
# Seasons for the data -
# spring 
# Summer
# fall
# winter 

df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3.head(5)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5  else 'Spring' if x > 2 else 'Winter')
```

```python
df3.head(3)
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```

```python
df3.query('race_length == "50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```
